```python
import random
import math
import statistics as stats
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
 def coin_flip():
    toss = random.choice([0, 1])
    
    if toss == 1:
        return True   #heads
    else:
        return False  #tails
    
```

```python
def repeat_flips(n = 2**16):
    total_flips = 2
    tails = 1
    heads = 1
    results = []
    for flip in range(n):
        if coin_flip() == True:
            heads += 1
            total_flips += 1
        else:
            tails += 1
            total_flips += 1
        if math.log2(flip+1) % 1 == 0:
            results.append((heads, tails))
    if sum(results [ len(results)-1 ]) != total_flips:
        results.append((heads, tails))
    return results
    
```

```python
plt.plot([x-y for x,y in repeat_flips()])
plt.ylabel('Difference of heads and tails.\n Negative numbers mean more tails,\n positive numbers mean more heads.')
plt.xlabel('Number of coin flips represented as 2 to the x power')
plt.title('Difference in Coin Flip Results')
plt.show
```

```python
# Somewhere around 2<sup>6</sup> the difference starts to veer away from 0 or equal numbers of heads and tails.
```

```python
plt.plot([x/y for x,y in repeat_flips()])
plt.title('Ratio of Heads to Tails Results: Trial 1')
plt.xlabel('Number of coin flips represented as 2 to the x power')
plt.show()
```

```python
plt.plot([x/y for x,y in repeat_flips()])
plt.title('Ratio of Heads to Tails Results: Trial 2')
plt.xlabel('Number of coin flips represented as 2 to the x power')
plt.show()
```

```python
plt.plot([x/y for x,y in repeat_flips()])
plt.title('Ratio of Heads to Tails Results: Trial 3')
plt.xlabel('Number of coin flips represented as 2 to the x power')
plt.show()
```

```python
# as the sample size approaches 2<sup>12</sup> the ratio of heads to tails reaches an equlibirum point near 1, or nearly the same numbers of heads as tails.
```
